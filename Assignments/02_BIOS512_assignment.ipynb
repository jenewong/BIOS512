Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

```r
library("scales")
```

```r
options(repr.plot.width=5, repr.plot.height=4, repr.plot.res = 300)
p = ggplot(penguins, aes(x = body_mass_g))
p = p + geom_histogram(fill = "cadetblue1", color = "cadetblue", bins = 37, alpha = 0.7)
p = p + facet_wrap(~year, nrow = 3, scale = "free_y")
p = p + labs(title = "Histogram of penguins body mass for measurements taken from 2007-2009", 
            y = "Number of penguins", x = "Body mass in grams")
p = p + theme_minimal() 
p = p + theme(plot.title = element_text(size = 8)) 
p = p + scale_x_continuous(labels = comma)
p = p + scale_y_continuous(breaks = c(0, 2.5, 5.0, 7.5))
p
```

Is there a difference between the body mass of male and female penguis? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to see compare within each species-group.**

```r
options(repr.plot.width=4, repr.plot.height=4, repr.plot.res = 300)
p2 = ggplot(penguins, aes(x = body_mass_g, y = sex))
p2 = p2 + geom_point(position = position_jitter(h = 0.25, w = 0.30), size = 1, alpha = 0.4)
p2 = p2 + facet_wrap(~species, nrow = 3)
p2 = p2 + labs(y = " ", x = "Body mass in grams")
p2 = p2 + theme_minimal()
p2 = p2 + scale_x_continuous(labels = comma, breaks = c(3500, 4500, 5500))
p2
```
