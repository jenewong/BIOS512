```r
library("tidyverse")
```

```r
wind = read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-10-27/wind-turbine.csv',
                  col_types = cols())
```

```r
wind %>% head
```

Each row in the table represents a single turbine.

Make a bar chart ðŸ“Š **sorted by frequency** of turbine count per province territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
wind_sorted = wind %>% 
    mutate(province_territory_sorted = province_territory %>% fct_infreq() %>% fct_rev()) 

p = ggplot(wind_sorted, aes(y = province_territory_sorted))
p = p + geom_bar(fill = "skyblue4") 
p = p + theme_minimal()
p = p + labs(y = "Province territory of turbine", 
             x = "Count", 
             title = "Turbine count per province territory")
p
```

Make a bar **sorted by frequency** of turbine count per manufacturer. **Group the low frequency manufacturers into an `Other` category**.

**Bonus:** Can you force the `Other` category to be the last bar? 

```r
wind_lumped = wind %>%
    mutate(manufacturer_lumped = manufacturer %>% 
           fct_lump(n = 6) %>% 
           fct_infreq() %>% 
           fct_relevel('Other', after = Inf) %>% 
           fct_rev())

p = ggplot(wind_lumped, aes(y = manufacturer_lumped))
p = p + geom_bar(fill = "skyblue4")
p = p + theme_minimal()
p = p + labs(y = "Manufacturer of turbine", 
             x = "Count",
            title = "Turbine count per manufacturer")
p
```
