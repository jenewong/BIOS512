Can you make this plot? ðŸ‘‡ What does this plot demonstrate?

![](img/patchwork_example_penguins.png)

ðŸ“Hint: add `scale_color_brewer(type = "qual", palette = "Set2")` to adjust the colors of the second plot!

```r
options(repr.plot.width=5, repr.plot.height=4)
p3 = ggplot(penguins, 
            aes(x = body_mass_g, 
                y = flipper_length_mm, 
                color = species))
p3 = p3 + geom_point(na.rm = TRUE)
p3
```

```r
options(repr.plot.width=5, repr.plot.height=4)
p4 = ggplot(penguins, 
            aes(x = body_mass_g, 
                y = flipper_length_mm, 
                color = island))
p4 = p4 + geom_point(na.rm = TRUE)
p4 = p4 + scale_color_brewer(type = "qual", palette = "Set2")
p4
```

```r
options(repr.plot.width=10, repr.plot.height=4)
p3 + p4

#interpretation: Most Gentoo penguins appear to come from Biscoe island
```

```r
#to combine the legends into one (comes from patchwork library)

options(repr.plot.width=10, repr.plot.height=4)
p3 + p4 + plot_layout(guides = "collect")
```

```r
options(repr.plot.width=10, repr.plot.height=10)
p5 = ggplot(penguins, 
            aes(x = body_mass_g, 
                y = flipper_length_mm))
p5 = p5 + geom_point(na.rm = TRUE)
p5 = p5 + facet_grid(island~species)
p5

#how to see what the species are in what island
```

# using stacked bars to show what birds on what island


think about if you have the data in the form you need 
y-axis is island and x-axis number (this is encoded in ggplot) 
color is species 

```r
options(repr.plot.width=5, repr.plot.height=3)
p6 = ggplot(penguins, 
            aes(y = island, fill = species)) 
    #can also flip around for y=species and fill=island
    #use fill instead of color to fill the bars instead of outlining the segments
    #don't have to define x-axis because ggplot2 will map it 
p6 = p6 + geom_bar(position = "fill") 
    #this will make a barchart 
p6 = p6 + scale_fill_brewer(type = "qual", palette = "Set2")
p6
    #geom_col could also produce the same chart 
```
